## Задача 1. Оценить месячный retention в оформление заказа с помощью когортного анализа.

```python
# исследуем датасет, чтоб определить по какому виду заказа будем считать retention
orders.groupby('order_status', as_index=False) \
      .agg(orders_count =('order_id','nunique')) \
      .sort_values(by='orders_count', ascending=False)
```

```python
# соединяем две таблицы с заказами и покупателями и оставляем только один вид заказа
merge_cust_ord = customers.merge(orders, how='left', on='customer_id').query('order_status == "delivered"')
merge_cust_ord
```

```python
# находим месячный и дневной периоды для каждой даты
merge_cust_ord['order_period'] = merge_cust_ord['order_purchase_timestamp'].dt.to_period('M')
merge_cust_ord['order_period_day'] = merge_cust_ord['order_purchase_timestamp'].dt.to_period('D')
```

```python
merge_cust_ord
```

```python
# оставляем нужные столбцы
merge_cust_ord = merge_cust_ord[['customer_id', 'customer_unique_id', 'order_id', 'order_status', 'order_purchase_timestamp', 'order_period', 'order_period_day']]
merge_cust_ord
```

```python
# находим первый месяц и день, когда покупатель первый раз познакомился с сервисом
merge_cust_ord['cohort'] = merge_cust_ord.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')
merge_cust_ord['cohort_day'] = merge_cust_ord.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('D')
```

```python
# узнаем номер месяц от даты заказа и первого дня
merge_cust_ord['period_number_month'] = np.floor((merge_cust_ord.order_period_day - merge_cust_ord.cohort_day) \
                                          .apply(attrgetter('n'))/30)
```

```python
merge_cust_ord
```

```python
# считаем количество человек на каждый период
df_cohort = merge_cust_ord.groupby(['cohort', 'period_number_month']) \
                          .agg(n_customers = ('customer_unique_id', 'nunique')) \
                          .reset_index()
df_cohort
```

```python
# переворачиваем таблицу и проводим когортный анализ
cohort_pivot = df_cohort.pivot_table(index='cohort', columns = 'period_number_month', values = 'n_customers')
cohort_pivot
```

```python
# находим нулевой период для определения retention
cohort_size = cohort_pivot.iloc[:, 0]
cohort_size
```

```python
# создаем месячный retention
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
retention_matrix
```

```python
# визуализируем данные по retention
with sns.axes_style('white'):
    fig, ax = plt.subplots(1,2,figsize=(16,12), sharey=True, gridspec_kw={'width_ratios': [1,11]})
    
    sns.heatmap(retention_matrix, mask=retention_matrix.isnull(), annot=True,fmt='.0%',cmap='RdYlGn',ax=ax[1])
    ax[1].set_title('Monthly Retention', fontsize=16)
    ax[1].set(xlabel='№ периода', ylabel='Когорта')
    
    cohort_size_df=pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, cbar=False, annot=True,fmt='g',cmap=white_cmap,ax=ax[0])
    
    fig.tight_layout()
    plt.show()
```

```python
# находим медианный retention 1-ого месяца
retention_matrix[1].describe()
# 50% квантиль явдяается медианной, для 1-го месяца это значение будет 0.003546
```

```python
# находим когорту с самым высоким retention за 3-й месяц
third_month = retention_matrix[3]
max_cohort  = third_month.idxmax()
max_value   = third_month.max().round(4)
print(f"Когорта с максимальным значением за 3-й месяц: {max_cohort}, значение: {max_value}")
```

## Задача 6. Подготовить дизайн эксперимента, зафиксировать все нужные показатели.

Proxy метрика: увеличение конверсии в повторные заказы\
Правила расчета метрики: отношения числа повторных заказов к числу всех заказов за определенный период\
Характер распределения: биноминальное (успех/неудача)\
Guardrail метрика: среднее значение времени доставки для всех заказов за период\
Нулевая гипотеза: Исправление бага не влияет на конверсию\

Ошибка второго рода: 20%\

Историческая дисперсия метрики: 8096622.514 \
MDE: 1%\
Минимальный размер выборки: 27617 

```python
#Добавим столбец с датой последней покупки для каждого пользователя
all_tabs_last = all_tabs.merge(all_tabs.groupby('customer_unique_id')['order_purchase_timestamp'].max().reset_index(), on='customer_unique_id', suffixes=('', '_last'))
all_tabs_last.head()
```

```python
all_tabs_last_copy = all_tabs_last.copy()
```

```python
delivered_orders = all_tabs_last_copy[all_tabs_last_copy['order_status'] == 'delivered']
```

```python
monthly_delivered = delivered_orders.groupby('period')['order_id'].count()
monthly_delivered
```

```python
historical_median = monthly_delivered.median()
print(round(historical_median, 3))
```

```python
historical_variance = monthly_delivered.var()
print(round(historical_variance, 3))
```

```python
import math

alpha = 0.05
beta = 0.2
z_alpha = st.norm.ppf(1 - alpha / 2)
z_beta = st.norm.ppf(1 - beta)

# Дисперсия
std_dev = math.sqrt(historical_variance)
mde = historical_median * 0.01

sample_size = ((z_alpha + z_beta) ** 2 * std_dev ** 2) / mde ** 2
print(round(sample_size))
```

```python
all_tabs_last_copy['group'] = np.random.choice(['control', 'test'], size=len(all_tabs_last_copy), p=[0.5, 0.5])
```

```python
all_tabs_last_copy.head()
```

```python
control = all_tabs_last_copy[all_tabs_last_copy['group'] == 'control']
test = all_tabs_last_copy[all_tabs_last_copy['group'] == 'test']

control_metric = control[control['order_status'] == 'delivered'].groupby('period')['order_id'].count()
test_metric = test[test['order_status'] == 'delivered'].groupby('period')['order_id'].count()
```

```python
control_metric.head()
```

```python
test_metric.head()
```

```python
from scipy.stats import ttest_ind

stat, p = ttest_ind(control_metric, test_metric)
print('t-statistic:', stat)
print('p-value:', p)
```

## Вывод - по p-value можно сказать, что у групп нет статистически значимых различий.
